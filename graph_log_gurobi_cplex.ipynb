```python
from extract_cplex import ExtractCPLEX
import pandas as pd
import matplotlib.pyplot as plt
import gurobi_logtools as glt
```

```python
results = glt.parse(["log_gurobi_perdidas"])
summary = results.summary()
dfg = results.progress("nodelog")
```

```python
cplex = ExtractCPLEX()
dfx, dic = cplex.mi_funcion_extract('log_cplex_AUHE_perdidas')
dfx['Time'] = dfx['seconds']
dfx['Gap'] = dfx['gap']
dfx['BestBd'] = dfx['BestBound']
dfx['Incumbent'] = dfx['bestInteger']
```

```python
plt.style.use('ggplot')
# plt.xlim(200, max(df["Time"]))
plt.plot(dfg["Time"], dfg["BestBd"], marker='o', linestyle='-',color='red')
plt.plot(dfg["Time"], dfg["Incumbent"], marker='o', linestyle='-',color='red')
plt.plot(dfx["Time"], dfx["BestBd"], marker='o', linestyle='-',color='blue')
plt.plot(dfx["Time"], dfx["Incumbent"], marker='o', linestyle='-',color='blue')
plt.xlabel('Segundos')
plt.ylabel('Funci√≥n objetivo')
plt.title('Convergencia CPLEX vs Gurobi')
plt.legend()  
plt.show()
```

```python
plt.close()
plt.style.use('ggplot')
# plt.xlim(200, max(df["Time"]))
plt.plot(dfg["Time"], dfg["Gap"], marker='o', linestyle='-',color='red')
plt.plot(dfx["Time"], dfx["Gap"], marker='o', linestyle='-',color='blue')
plt.xlabel('Segundos')
plt.ylabel('Gap')
plt.title('Convergencia CPLEX vs Gurobi (GAP)')
plt.legend()  
plt.show()
```
